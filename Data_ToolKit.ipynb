Que.2  How does broadcasting work in NumPy

Ans.The term broadcasting describes how NumPy treats arrays with different shapes during arithmetic operations. Subject to certain constraints, the smaller array is “broadcast” across the larger array so that they have compatible shapes. Broadcasting provides a means of vectorizing array operations so that looping occurs in C instead of Python. It does this without making needless copies of data and usually leads to efficient algorithm implementations. There are, however, cases where broadcasting is a bad idea because it leads to inefficient use of memory that slows computation.

NumPy operations are usually done on pairs of arrays on an element-by-element basis. In the simplest case, the two arrays must have exactly the same shape, as in the following example:

import numpy as np
a = np.array([1.0, 2.0, 3.0])
b = np.array([2.0, 2.0, 2.0])
a * b
array([2.,  4.,  6.])

Que.10 What is the significance of hierarchical indexing in Pandas

Ans. Hierarchical indexing, also known as multi-indexing, is a feature of pandas that allows you to have multiple index levels on an axis. It provides a way to work with higher-dimensional data using the familiar one-dimensional Series and two-dimensional DataFrame structures.

Creating a MultiIndex

You can create a MultiIndex in pandas using various constructors provided by the pd.MultiIndex class. For instance, you can use from_arrays, from_tuples, from_product, or directly by using the MultiIndex constructor with levels and labels.

Reshaping with MultiIndex

Pandas provides methods like stack and unstack which can help in reshaping the data from a stacked multi-index to a simple two-dimensional representation and vice versa. For example, to unstack the second index level.

MultiIndex for Columns:

MultiIndex can also be used for columns in a DataFrame, allowing for more complex data representations. For example, you can have a DataFrame with patient data indexed by both the year and the visit number, and with columns indexed by both the patient name and the measurement type.

Data Aggregations on Multi-Indices

Pandas' built-in data aggregation methods, such as mean(), sum(), and max(), can be used with a level parameter to control the subset of data the aggregate is computed on. This allows for flexible and powerful data analysis workflows.

Converting Between Simple and Hierarchical Index

You can convert between a simple index and a hierarchical index using the reset_index and set_index methods. This can be particularly useful when dealing with real-world datasets that start with a simple index.

Que.15 How does NumPy handle multidimensional arrays)

Ans.NumPy is a fundamental package for scientific computing in Python, providing support for multi-dimensional arrays and matrices, along with a collection of mathematical functions to operate on these arrays. The core feature of NumPy is the ndarray object, which is a multi-dimensional array of fixed-size items, allowing for efficient array computations.

Creating NumPy Arrays

NumPy arrays, or ndarrays, can be created through various methods. The most common way is to use the np.array function, which can take a list or a list of lists as input and convert it into an ndarray. The shape of the array is determined by the number of elements in each dimension, and the type of elements is inferred or can be explicitly specified using the dtype parameter.

Accessing Array Elements

Elements in a NumPy array can be accessed using indexing and slicing, similar to Python lists. For multi-dimensional arrays, you can use a comma-separated tuple of indices to access elements. Slicing can also be used to access a subset of the array.

Advanced Indexing

NumPy provides advanced indexing techniques, such as boolean indexing and fancy indexing, which allow for more complex data manipulation.

Memory Layout

The internal memory layout of an ndarray is defined by its shape, strides, and data type. NumPy arrays can be stored in row-major (C-style) or column-major (Fortran-style) order, affecting how the elements are laid out in memory.

Que.2  Write a Pandas script to find the mean of a specific column in a DataFrame

```python
import pandas as pd
data = {
    'A': [1, 2, 3, 4, 5],
    'B': [10, 20, 30, 40, 50],
    'C': [100, 200, 300, 400, 500]
}

df = pd.DataFrame(data)

# Specify the column for which you want to find the mean
column_name = 'B'

# Calculate the mean of the specified column
mean_value = df[column_name].mean()

print(f"The mean of column '{column_name}' is {mean_value}")
```

Que. Create a 3D scatter plot using Plotly

```python
import plotly.graph_objects as go
x = [1, 2, 3, 4, 5]
y = [10, 11, 12, 13, 14]
z = [5, 6, 7, 8, 9]
fig = go.Figure(data=[go.Scatter3d(
    x=x,
    y=y,
    z=z,
    mode='markers',
    marker=dict(
        size=5,
        color=z,
        colorscale='Viridis',
        opacity=0.8
    )
)])
fig.update_layout(
    title='3D Scatter Plot',
    scene=dict(
        xaxis_title='X Axis',
        yaxis_title='Y Axis',
        zaxis_title='Z Axis'
    )
)
fig.show()
```
